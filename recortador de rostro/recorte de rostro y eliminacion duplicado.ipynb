Recorte de rosotros y eliminacion de duplicados en el dataset 2

```python
import os
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def recortar_rostros(dataset):
    print('Recortando rostros...')
    filelist = os.listdir(dataset)
    n_imgs = len(filelist)
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt2.xml') 
    
    for i in range(n_imgs): #n_imgs
        print("Imagen", i, "de", n_imgs)
        direcion = str(dataset)+ "\\" + str(filelist[i])
        
        imagen = cv2.imread(direcion)
        gray = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
        
        faces = face_cascade.detectMultiScale(gray, 1.1, 5)
            
        for (x, y, w, h) in faces:
            cv2.rectangle(imagen,(x, y),(x+w, y+h),(0, 0, 255), 2)
            faces = imagen[y:y + h, x:x + w] 
            cv2.imwrite('dataset2/soloRostros/face'+str(i)+'.jpg', faces) 

    print('Â¡Listo!')
```

```python
def comparando_img(dataset, salto = 0):
    print('Comparando rostros...')
    filelist = os.listdir(dataset)
    n_imgs = len(filelist)

    for i in range(salto, n_imgs):
        print("Imagen", i, "de", n_imgs, "")
        direcion = str(dataset)+ "\\" + str(filelist[i])
        
        imagen = cv2.imread(direcion)
        
        filelist_sinRepetir = os.listdir("dataset2\sinrepetir")
        n_imgs_sinRepetir = len(filelist_sinRepetir)
        
        for j in range(n_imgs_sinRepetir):
            
            direcion_sinRepetir = "dataset2\sinrepetir" + "\\" + str(filelist_sinRepetir[j])      
            imagen_sinRepetir = cv2.imread(direcion_sinRepetir)
            
            if imagen.shape != imagen_sinRepetir.shape:
                continue
            
            difference = cv2.subtract(imagen, imagen_sinRepetir)
            
            if np.any(difference):
                print("repedido")
                break
        else:
            cv2.imwrite('dataset2/sinrepetir/face'+str(i)+'.jpg', imagen)    
        
    print('Â¡Listo!')
```
